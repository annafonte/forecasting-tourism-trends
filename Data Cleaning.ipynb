```python
df = pd.read_csv("../forecasting-tourism-trends/tourism-data.csv", error_bad_lines=False)
```

#### Inspecting the dataset

```python
#Keeping the desired columns
df = df[["Temps", "País de residència", "Valors"]]

#Renaming the columns
df.columns = ["period", "country", "value"]
```

```python
#Arranging types of columns: value to float and period to datetime
df["value"] = df["value"].replace({':': np.nan}).astype(float) #Replacing : to nulls for dealing with them later
df["period"] = pd.to_datetime(df["period"], format="%m/%Y")
df.dtypes
```

```python
# Creating pivot table
df = df.pivot(index='period', columns='country', values='value').reset_index()
df.head(3)
```

```python
# Calculating summing row to work with top 5 countries
df.loc['Total',:]= df.sum(axis=0)
```

```python
# The top 5 countries that visit Catalonia are France, United Kingdom, Germany, United States and Italy 
df.sort_values(by ='Total', axis=1, ascending=False).tail(1)
```

```python
# Creating a subset df with
df_top5 = df[["period", "França", "Regne Unit", "Alemanya", "Estats Units", "Itàlia", "Total"]].copy()
```
